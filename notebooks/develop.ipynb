## agents

```python
msg = "where can i find info about accounts?"
user_input = HumanMessage(content=msg)
```

### planner

```python
PLANNER_TOOLS = [list_data_sources]
planner_agent = create_agent(planner_llm, tools=PLANNER_TOOLS)

SYSTEM_MSG = """
You are an expert analystics engineer with a specialty in translating a business partner's questions into a series of well defined steps to gather the data needed to answer the quesion. Use the data catalog tools to help you understand what data is available & how it might help answer the question.
"""
sys_msg = SystemMessage(content=SYSTEM_MSG)


prompt = [
    sys_msg,
    user_input,
]
response = planner_agent.invoke({"messages": prompt})
print(response)
```
